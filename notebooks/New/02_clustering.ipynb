# 02 · Dimensionality reduction & clustering
**Input :** `data/processed/adata_processed.h5ad`  
**Output:** `data/processed/adata_clustered.h5ad`

**Run order:** 01 → **02** → 03 → 04 → 05


```
import sys
from pathlib import Path

def _find_repo_root(start):
    for p in [start, *start.parents]:
        if (p / "paths.py").exists():
            return p
    raise FileNotFoundError("paths.py not found — run: python scripts/data_download.py")

REPO_ROOT = _find_repo_root(Path.cwd())
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))
if str(REPO_ROOT / "scripts") not in sys.path:
    sys.path.insert(0, str(REPO_ROOT / "scripts"))

from paths import REPO_ROOT, RAW_DIR, PROC_DIR, FIGURES_DIR, TABLES_DIR, MODELS_DIR
print(f"Repo root : {REPO_ROOT}")
print(f"Raw data  : {RAW_DIR}")
```

```
import scanpy as sc
from scrna_functions import run_pca, run_umap, run_leiden, save_adata
```

```
# ── Configuration ─────────────────────────────────────────────────────────────
N_NEIGHBORS  = 15
N_PCS        = 10
RESOLUTIONS  = (0.3, 0.5, 1.0, 2.0)
```

## 1 · Load

```
adata = sc.read(str(PROC_DIR / "adata_processed.h5ad"))
print(adata)
```

## 2 · PCA

```
adata = run_pca(adata)
```

## 3 · UMAP

```
adata = run_umap(adata, n_neighbors=N_NEIGHBORS, n_pcs=N_PCS)
sc.pl.umap(adata, color="sample", size=2)
```

## 4 · Leiden clustering

```
adata = run_leiden(adata, resolutions=RESOLUTIONS)
sc.pl.umap(adata, color=["leiden_res_0.50","sample"], wspace=0.4, size=6)
```

## 5 · Save

```
save_adata(adata, PROC_DIR / 'adata_clustered.h5ad')
```
